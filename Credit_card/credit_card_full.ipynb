```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('/content/creditcard.csv')
df.head()
```

```python
df.info()
```

```python
df['Class'].value_counts()
```

```python
legit=df[df.Class==0.0]
fraud=df[df.Class==1.0]
```

```python
legit.shape, fraud.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
df.groupby('Class').mean()
```

#balancing the data using sampling

```python
legit_sample=legit.sample(n=52)
```

```python
new_df=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
x=new_df.drop(columns="Class", axis=1)
y=new_df['Class']
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2,stratify=y)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_prediction=model.predict(x_train)
training_data_acc=accuracy_score(x_train_prediction,y_train)
```

```python
print(training_data_acc)
```

```python
x_test_prediction=model.predict(x_test)
testing_data_acc=accuracy_score(x_test_prediction,y_test)
```

```python
print(testing_data_acc)
```
